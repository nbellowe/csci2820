The main goal of this lecture is to introduce the powerful LU-Decomposition of a matrix.  It is the first practical implementation of Gaussian Elimination that we've seen in the course that is actually used in computer science practice.  Along the way we're going to introduce two new concepts: the matrix **inverse** and the matrix **transpose**.  We'll talk at great length later in the week and the course about these topics, but for now we'll motivate them by their application to the LU-Decomposition. 

Consider the example from last time where we performed Gaussian Elimination on an augmented system, and kept track of the elimination matrices along the way.  This time we'll forget the right-hand side vector (which is kinda the point of the LU Decomposition).  

###Example 1 

Use Gaussian Elimination to reduce the following matrix to upper triangular form: $~~~~A = \mymat{2 & -1 & 3 \\ 4 & 0 & 7 \\ 2 & 1 & 8}$. 

<br>

$$
\mymat{
2 & -1 & 3 \\
4 & 0  & 7 \\
2 & 1 & 8 
}
\quad \quad 
\begin{array}{ll}
& \\
R_2 \leftarrow & R_2 - 2R_1 \\
&
\end{array}
\quad \quad
\mymat{
2 & -1 & 3 \\
0 & 2  & 1 \\
2 & 1 & 8 
}, 
\quad
E_{21} = 
\mymat{
1 & 0 & 0 \\
-2 & 1 & 0 \\
0 & 0 & 1 \\
}
$$

<br>

$$
\mymat{
2 & -1 & 3 \\
0 & 2  & 1 \\
2 & 1 & 8 
}
\quad \quad 
\begin{array}{ll}
& \\
& \\
R_3 \leftarrow & R_3 - R_1 
\end{array}
\quad \quad
\mymat{
2 & -1 & 3 \\
0 & 2  & 1 \\
0 & 2 & 5 
}, 
\quad
E_{31} = 
\mymat{
1 & 0 & 0 \\
0 & 1 & 0 \\
-1 & 0 & 1 \\
}
$$

<br>

$$
\mymat{
2 & -1 & 3 \\
0 & 2  & 1 \\
2 & 1 & 8 
}
\quad \quad 
\begin{array}{ll}
& \\
& \\
R_3 \leftarrow & R_3 - R_2 
\end{array}
\quad \quad
\mymat{
2 & -1 & 3 \\
0 & 2  & 1 \\
0 & 0 & 4 
}, 
\quad
E_{32} = 
\mymat{
1 & 0 & 0 \\
0 & 1 & 0 \\
0 & -1 & 1 \\
}
$$

<br>

So we have 

<br>

$$
E_{32}E_{31}E_{21}A = U = 
\mymat{
2 & -1 & 3 \\
0 & 2  & 1 \\
0 & 0 & 4 
}
$$
