##Systems of Three Equations and Three Unknowns

<br>

In the previous class we solved systems of two equations and two unknowns in a systematic way by transforming the linear system into an upper triangular system through **elimination** and then solving the new system by **back substitution**.  Let's look at how the process generalizes to systems of three equations and three unknowns.  After this we'll have the tools necessary to solve systems of any reasonable size. 

The goals here are still the same 

1. Perform **elimination** to make the system look upper triangular, either by adding multiples of one row to another, or possibly by exchanging rows. 
2. Solve the resulting upper triangular system via back substitution.

###Example 1 

Consider the following system 

<br>

$$
\begin{array}{rcl}
x_1 + 3x_2 + 2x_3 &=& 4 \\
-2x_1 - 7x_2 + x_3 &=& 9 \\
3x_1 + 10x_2 + 9x_3 &=& 19
\end{array}
$$

<br>

The first thing we'll do is use the **pivot** in the $x_1$ position of row 1 to eliminate the $x_1$ terms in the other equations. 

To eliminate the $x_1$ term in row 2 we multiply the first row by $-2$ and subtract it from row 2.  To eliminate the $x_1$ term in row 3 we multiply the first row by $3$ and subtract it from row 3. 

<br>

$$
\begin{array}{rcl}
x_1 + 3x_2 + 2x_3 &=& 4 \\
 -x_2 + 5x_3 &=& 17 \\
 x_2 + 3x_3 &=& 7
\end{array}
\quad \quad 
\begin{array}{ll}
 & \\
R_2 \leftarrow & R_2 - (-2)R_1  \\
R_3 \leftarrow & R_3 - 3R_1 
\end{array}
$$

<br>

We've successfully eliminated all of the $x_1$ terms in rows 2 and 3.  Notice that the 1 coefficient on $x_1$ is the **pivot** in the first row and we used it to eliminate the $x_1$ terms in rows 2 and 3 with **multipliers** of $-2$ and $3$, respectively.  Recall that the multiplier is computed by 

<br>

$$
\textbf{multiplier} = \textbf{entry to eliminate} \div \textbf{pivot}
$$

<br>

Next we want to eliminate the $x_2$ term in row 3.  Notice that we need to subtract a multiple of one of the previous rows to eliminate this term, but if we use the first equation this will put a nonzero $x_1$ term in row 3 and muck up our upper triangular form.  This means that we can really only use row 2 to do this elimination.  In fact, at this point we're pretty much done with row 1 for the elimination phase. 

The **pivot** in the second row is the $-1$ in front of the $x_2$ term.  To eliminate the $x_2$ in row three we multiply row 2 by $-1$ and subtract.  

<br>

$$
\begin{array}{rcl}
x_1 + 3x_2 + 2x_3 &=& 4 \\
 -x_2 + 5x_3 &=& 17 \\
 8x_3 &=& 24
\end{array}
\quad \quad 
\begin{array}{ll}
 & \\
& \\
R_3 \leftarrow & R_3 - (-1)R_2
\end{array}
$$

<br>

The system is now in upper triangular form and the elimination phase is done.  The new upper triangular system is 

<br>

$$
\begin{array}{rcl}
x_1 + 3x_2 + 2x_3 &=& 4 \\
 -x_2 + 5x_3 &=& 17 \\
 8x_3 &=& 24
\end{array}
$$

<br>

Notice that the system has three pivots, namely $1$, $-1$, and $8$.  To solve the system for the unknowns $x_1$, $x_2$, and $x_3$ we perform back substitution.  

Starting with row 3 we have 

<br>

$$
x_3 = 24 ~/~8 = 3 
$$

<br>

Substituting this into row 2 and solving for $x_2$ we have 

<br>

$$
-x_2 + 5\myp{3} = 17 \quad \Rightarrow \quad -x_2 = 2 \quad \Rightarrow \quad x_2 = 2 ~~ / ~ -1 = -2  
$$

<br>

Substituting the values for $x_3$ and $x_2$ into row 1 and solving for $x_1$ we have 

<br>

$$
x_1 + 3\myp{-2} + 2\myp{3} = 4 \quad \Rightarrow \quad x_1 = 4
$$

<br>

Which gives us a final solution of ${\bx} = \mymat{4 \\ -2 \\ 3}$

OK, things are starting to get awfully clunky solving the algebraic system with so many equations and unknowns.  It's time to make things nicer by formulating things with matrices and vectors.  Recall that the above system can be written in matrix equation form as 

<br>

$$
\mymat{
1 & 3 & 2 \\
-2 & -7 & 1 \\
3 & 10 & 9 
}
\mymat{x_1 \\ x_2 \\ x_3} = 
\mymat{4 \\ 9 \\ 19}
$$

<br>

The beauty of the matrix formulation is that we have all of the necessary coefficients of the linear system encoded in the matrix.  Their position in the matrix tells us which equation and unknown each coefficient corresponds to, i.e. each row of the matrix refers to an equation in the orginal system, and each column in the matrix corresponds to one of the unknown variables.  We should be able to perform exactly the same operations on the matrix and right-hand side vector that we performed on the system of equations.  

In fact, since we know which column corresponds to which unknown, that vector of variables is kind of in the way too.  Let's get rid of it. 

<br>

$$
\myaug{
1 & 3 & 2 & 4 \\
-2 & -7 & 1 & 9 \\
3 & 10 & 9 & 19
}
$$

<br>

The first three columns of the table represent the original coefficient matrix.  The last column represents the rhs vector.  The vertical line is there to tell us where the separation between matrix and rhs vector is.  

We can now proceed as before with the elimination phase.  Nothing has really changed except the way we have written down the linear system.  We choose the $1$ in the $(1,1)$ position of the coefficient matrix as the first pivot and use it to eliminate the nonzero coefficients beneath it in the first column.  We have 

<br>

$$
\myaug{
1 & 3 & 2 & 4 \\
-2 & -7 & 1 & 9 \\
3 & 10 & 9 & 19
}
\quad \quad
\begin{array}{l&l}
 & \\
R_2 \leftarrow & R_2 - (-2)R_1  \\
R_3 \leftarrow & R_3 - 3R_1 
\end{array}
\quad \quad
\myaug{
1 & 3 & 2 & 4 \\
0 & -1 & 5 & 17 \\
0 & 1 & 3 & 7
}
$$

$$
\myaug{
1 & 3 & 2 & 4 \\
0 & -1 & 5 & 17 \\
0 & 1 & 3 & 7
}
\quad \quad
\begin{array}{l&l}
 & \\
 & \\
 R_3 \leftarrow & R_3 - (-1)R_2 
\end{array}
\quad \quad
\myaug{
1 & 3 & 2 & 4 \\
0 & -1 & 5 & 17 \\
0 & 0 & 8 & 24
}
$$

<br>

which gives us the upper triangular augmented system 

<br>

$$
\myaug{
1 & 3 & 2 & 4 \\
0 & -1 & 5 & 17 \\
0 & 0 & 8 & 24
}
$$

<br>

Note that in this form it's even easier to identify the pivots $1$, $-1$, $8$.  They're sitting right there on the diagonal of the matrix. 

The back substitution phase can easily be carried out on the augmented system.  We typically write it in the following condensed form.  Note that although it's written in the usual order, you kind of have to read it from the bottom down. 

<br>

$$
\myaug{
1 & 3 & 2 & 4 \\
0 & -1 & 5 & 17 \\
0 & 0 & 8 & 24
}
\quad \Rightarrow \quad
\bx = 
\mymat{
(4 - 3(-2) - 2(3)) / 1 = 4 \\
(17 - 5(3)) / -1 = -2 \\
24 / 8 = 3 \\
} = 
\mymat{
4 \\
-2 \\
3
}
$$
